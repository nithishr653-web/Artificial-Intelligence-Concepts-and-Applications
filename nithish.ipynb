```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag,ne_chunk

text = "Apple is based in california"
tokens = word_tokenize(text)
pos_tag = pos_tag(tokens)

ner = ne_chunk(pos_tag)
print(ner)
```
